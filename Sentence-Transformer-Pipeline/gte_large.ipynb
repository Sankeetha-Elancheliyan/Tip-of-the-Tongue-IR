```python
import os
import json
import torch
import faiss
import numpy as np
import pandas as pd
import pickle
from tqdm import tqdm
from sentence_transformers import SentenceTransformer, losses
import logging
import gc
from torch.utils.data import DataLoader, Dataset
```

```python
# Load GTE-Large model
device = "cuda" if torch.cuda.is_available() else "cpu"
log.info(f"Loading model on {device}")
# Load GTE-Large model
model = SentenceTransformer("Alibaba-NLP/gte-large-en-v1.5", device=device, trust_remote_code=True) # Add trust_remote_code=True
```

```python



#  Convert JSONL to CSV

def convert_jsonl_to_csv(jsonl_path, csv_path):
    """Converts JSONL to CSV for faster processing."""
    log.info("Converting JSONL to CSV...")

    # Read JSONL line by line and store in a list
    data = []
    with open(jsonl_path, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))

    # Convert to DataFrame
    df = pd.DataFrame(data)

    # Save as CSV
    df.to_csv(csv_path, index=False, encoding='utf-8')
    log.info(f"Conversion complete! CSV saved at: {csv_path}")

# Check if CSV exists, otherwise create it
if not os.path.exists(CORPUS_CSV):
    convert_jsonl_to_csv(CORPUS_JSONL, CORPUS_CSV)
else:
    log.info("CSV already exists, skipping conversion.")


```

```python
#  Load CSV into DataFrame

def load_csv_to_df(csv_path):
    """Loads CSV file into a Pandas DataFrame."""
    log.info("Loading CSV into DataFrame...")
    df = pd.read_csv(csv_path, encoding='utf-8')
    log.info(f"Loaded {len(df)} documents successfully.")
    return df

# Load the dataset
df = load_csv_to_df(CORPUS_CSV)

# Display first few rows
df.head()
```

```python
# 1 Load JSONL File as DataFrame (Line-by-Line Processing)

def load_jsonl_to_df(file_path):
    """Loads JSONL file line-by-line into a Pandas DataFrame."""
    data = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return pd.DataFrame(data)

log.info("Loading corpus into DataFrame...")
df = load_jsonl_to_df(CORPUS_PATH)
log.info(f"Loaded {len(df)} documents.")
```

```python
#  2️ FAISS Indexing
def create_faiss_index(df, model, batch_size=16):
    """Encodes text and indexes them using FAISS without converting to lists."""
    index = None
    for i in tqdm(range(0, len(df), batch_size), desc="Encoding & Indexing"):
        batch_texts = df.iloc[i:i + batch_size]["text"].values  # Access batch from DataFrame
        batch_ids = df.iloc[i:i + batch_size]["doc_id"].values  # Use batch IDs

        with torch.no_grad():
            batch_embeddings = model.encode(batch_texts, convert_to_tensor=False)

        if index is None:
            d = batch_embeddings.shape[1]
            index = faiss.IndexFlatL2(d)

        index.add(batch_embeddings)

        # Free memory
        torch.cuda.empty_cache()
        gc.collect()

    return index, df  

try:
    with open(FAISS_INDEX_FILE, "rb") as f:
        faiss_index, stored_df = pickle.load(f)
    log.info("Loaded FAISS index from file.")
except FileNotFoundError:
    log.info("FAISS index not found. Creating a new one...")
    faiss_index, df = create_faiss_index(df, model, batch_size=16)

    with open(FAISS_INDEX_FILE, "wb") as f:
        pickle.dump((faiss_index, df), f)
    log.info("FAISS index saved.")

log.info("FAISS index is ready.")
```

```python
#  3️ Query Retrieval & Evaluation (Before Fine-Tuning)

def retrieve_top_k(df, query_texts, top_k=10):
    """Retrieve top-K documents using FAISS without `.tolist()`."""
    results = {}

    for query_id, query_text in tqdm(query_texts.items(), desc="Retrieving Queries"):
        query_embedding = model.encode([query_text], convert_to_tensor=False).reshape(1, -1)
        distances, indices = faiss_index.search(query_embedding, top_k)

        retrieved_docs = {
            df.iloc[idx]["doc_id"]: float(dist) for idx, dist in zip(indices[0], distances[0])
        }

        results[query_id] = retrieved_docs

    return results
```

```python
# Load Queries & Qrels
def load_queries(query_file):
    queries = {}
    with open(query_file, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            queries[data["query_id"]] = data["text"]
    return queries

def load_qrels(qrel_file):
    qrels = {}
    with open(qrel_file, 'r', encoding='utf-8') as f:
        for line in f:
            qid, _, docid, rel = line.strip().split()
            if qid not in qrels:
                qrels[qid] = {}
            qrels[qid][docid] = int(rel)
    return qrels

train_queries = load_queries(TRAIN_QUERIES_PATH)
train_qrels = load_qrels(TRAIN_QREL_PATH)
test_queries = load_queries(TEST_QUERIES_PATH)
```

```python
# Retrieve before fine-tuning
before_finetune_results = retrieve_top_k(df, test_queries)

import pytrec_eval

def evaluate_retrieval(retrieved_results, qrels):
    """Evaluates retrieval performance."""
    evaluator = pytrec_eval.RelevanceEvaluator(qrels, {'ndcg', 'map', 'recall', 'recip_rank'})
    eval_results = evaluator.evaluate(retrieved_results)

    aggregated_results = {
        metric: np.mean([res[metric] for res in eval_results.values()])
        for metric in eval_results[list(eval_results.keys())[0]].keys()
    }

    for metric, value in aggregated_results.items():
        log.info(f"{metric}: {value:.4f}")

    return aggregated_results

before_finetune_eval = evaluate_retrieval(before_finetune_results, train_qrels)
```

```python
# 4️ Fine-Tune Model

class TrainDataset(Dataset):
    """Custom Dataset for Training SBERT."""
    def __init__(self, queries, qrels):
        self.data = [(queries[qid], docid) for qid, doc_dict in qrels.items() for docid in doc_dict]

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

train_dataset = TrainDataset(train_queries, train_qrels)
train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=16)

train_loss = losses.MultipleNegativesRankingLoss(model)

log.info("Fine-tuning model...")
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=3, warmup_steps=100)

model.save(MODEL_DIR)
log.info("Fine-tuning complete!")

# Retrieve After Fine-Tuning
after_finetune_results = retrieve_top_k(df, test_queries)
after_finetune_eval = evaluate_retrieval(after_finetune_results, train_qrels)
```

```python
# Save Evaluation Results

RESULTS_FILE = os.path.join(DATA_PATH, "evaluation_results.json")
with open(RESULTS_FILE, "w") as f:
    json.dump({"before_finetune": before_finetune_eval, "after_finetune": after_finetune_eval}, f)

log.info("All results saved!")
```

```python
#  1️ FAISS Indexing (Streaming, No DataFrame)

def create_faiss_index(jsonl_file, model, batch_size=64):
    """Indexes text using FAISS without loading the full dataset into memory."""
    index = None

    with open(jsonl_file, 'r', encoding='utf-8') as f:
        # Estimate total lines for tqdm (to show progress)
        total_lines = sum(1 for _ in open(jsonl_file))  # Get an estimate of the total lines
        f.seek(0)  # Reset file pointer to the beginning

        # Loop through batches with tqdm progress bar
        for i in tqdm(range(0, total_lines, batch_size), desc="Encoding & Indexing", total=total_lines // batch_size + 1):
            batch = []
            doc_ids = []

            # Read batch from JSONL
            for _ in range(batch_size):
                line = f.readline()
                if not line:
                    break  # End of file
                doc = json.loads(line)
                batch.append(doc["text"])
                doc_ids.append(doc["doc_id"])

            if not batch:
                break  # No more data to process

            # Encode batch
            with torch.no_grad():
                # Reduce batch size for encoding to further mitigate OOM
                batch_embeddings = []
                for text in batch:
                    embedding = model.encode([text], convert_to_tensor=False)
                    batch_embeddings.append(embedding)
                batch_embeddings = np.vstack(batch_embeddings)

            if index is None:
                d = batch_embeddings.shape[1]
                index = faiss.IndexFlatL2(d)

            index.add(batch_embeddings)  # Add to FAISS index

            # Aggressive memory management within the loop
            del batch_embeddings
            torch.cuda.empty_cache()
            gc.collect()

    return index
```

```python
#  1️ FAISS Indexing (Streaming, No DataFrame)

def create_faiss_index(jsonl_file, model, batch_size=64):
    """Indexes text using FAISS without loading the full dataset into memory."""
    index = None

    with open(jsonl_file, 'r', encoding='utf-8') as f:
        for i in tqdm(range(0, 10**10, batch_size), desc="Encoding & Indexing"):
            batch = []
            doc_ids = []

            # Read batch from JSONL
            for _ in range(batch_size):
                line = f.readline()
                if not line:
                    break  # End of file
                doc = json.loads(line)
                batch.append(doc["text"])
                doc_ids.append(doc["doc_id"])

            if not batch:
                break  # No more data to process

            # Encode batch
            with torch.no_grad():
                batch_embeddings = model.encode(batch, convert_to_tensor=False)

            if index is None:
                d = batch_embeddings.shape[1]
                index = faiss.IndexFlatL2(d)

            index.add(batch_embeddings)  # Add to FAISS index

            torch.cuda.empty_cache()
            gc.collect()

    return index
```

```python
#  2️ FAISS Query Retrieval (Streaming, No Full Load)

def retrieve_faiss(jsonl_file, queries, batch_size=32, top_k=10, n_threads=4):
    """Retrieves top-K documents using FAISS efficiently without loading full dataset."""
    faiss.omp_set_num_threads(n_threads)  # Use multi-threading
    run = {}

    for query_id, query_text in tqdm(queries.items(), desc="Retrieving Queries"):
        with torch.no_grad():
            query_embedding = model.encode([query_text], convert_to_tensor=False)

        distances, indices = faiss_index.search(query_embedding, top_k)  # FAISS batch search
        retrieved_docs = {}

        # Read JSONL line-by-line to find doc IDs for retrieved indices
        with open(jsonl_file, 'r', encoding='utf-8') as f:
            for i, line in enumerate(f):
                if i in indices[0]:  # Only read necessary lines
                    doc = json.loads(line)
                    retrieved_docs[doc["doc_id"]] = float(distances[0][list(indices[0]).index(i)])

        run[query_id] = retrieved_docs

    return run
```

```python
# Load Queries (No Full Dataset Load)
def load_queries(query_file):
    with open(query_file, 'r', encoding='utf-8') as f:
        return {json.loads(line)["query_id"]: json.loads(line)["text"] for line in f}

def load_qrels(qrel_file):
    qrels = {}
    with open(qrel_file, 'r', encoding='utf-8') as f:
        for line in f:
            qid, _, docid, rel = line.strip().split()
            if qid not in qrels:
                qrels[qid] = {}
            qrels[qid][docid] = int(rel)
    return qrels
```

```python
TRAIN_QUERIES_PATH = os.path.join(DATA_PATH, 'train-2024/queries.jsonl')
TRAIN_QREL_PATH = os.path.join(DATA_PATH, 'train-2024/qrel.txt')
TEST_QUERIES_PATH = os.path.join(DATA_PATH, 'test-2024/queries.jsonl')

train_queries = load_queries(TRAIN_QUERIES_PATH)
train_qrels = load_qrels(TRAIN_QREL_PATH)
test_queries = load_queries(TEST_QUERIES_PATH)
```

```python
# Retrieve before fine-tuning
before_finetune_results = retrieve_faiss(CORPUS_PATH, test_queries)

import pytrec_eval

def evaluate_retrieval(retrieved_results, qrels):
    """Evaluates retrieval performance."""
    evaluator = pytrec_eval.RelevanceEvaluator(qrels, {'ndcg', 'map', 'recall', 'recip_rank'})
    eval_results = evaluator.evaluate(retrieved_results)

    aggregated_results = {
        metric: np.mean([res[metric] for res in eval_results.values()])
        for metric in eval_results[list(eval_results.keys())[0]].keys()
    }

    for metric, value in aggregated_results.items():
        log.info(f"{metric}: {value:.4f}")

    return aggregated_results

before_finetune_eval = evaluate_retrieval(before_finetune_results, train_qrels)
```

```python
# 3️ Fine-Tune GTE-Large

from torch.utils.data import DataLoader, Dataset
from sentence_transformers import losses

class TrainDataset(Dataset):
    """Custom Dataset for Training SBERT."""
    def __init__(self, queries, qrels):
        self.data = [(queries[qid], docid) for qid, doc_dict in qrels.items() for docid in doc_dict]

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

train_dataset = TrainDataset(train_queries, train_qrels)
train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=16)

train_loss = losses.MultipleNegativesRankingLoss(model)

log.info("Fine-tuning model...")
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=3, warmup_steps=100)

model.save(MODEL_DIR)
log.info("Fine-tuning complete!")

# Retrieve After Fine-Tuning
after_finetune_results = retrieve_faiss(CORPUS_PATH, test_queries)
after_finetune_eval = evaluate_retrieval(after_finetune_results, train_qrels)
```
